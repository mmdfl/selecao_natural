```python
import random

def pega_dois_individuos_aleatorios(pop):
    #
    casal=random.sample(pop,2)
    #verificar se podem cruzar
    #se  não podem, retorna []
    #se podem:
    return casal


def gerar_filhos(casal):
    #crossing_over
    fator_de_individualidade=(0.8+0.4*random.random())
    novo_individuo=fator_de_individualidade*(casal[0]+casal[1])/2
    return novo_individuo

def add_filho_a_pop(filho, pop):
    pop.append(filho)
    return pop

# casal_teste=pega_dois_individuos_aleatorios(pop)
# print(casal_teste)
# filho_teste=gerar_filhos(casal_teste)
# print(filho_teste)
```

```python
def cruzar_aleatoriamente(pop):
    qtde_tentativas_de_cruzamentos=random.randint(0,10)
    i=0
    while i<qtde_tentativas_de_cruzamentos:
        casal=pega_dois_individuos_aleatorios(pop)
        filho=gerar_filhos(casal)
        pop=add_filho_a_pop(filho, pop)
        i=i+1
    return pop

# print(pop)
# pop=cruzamentos_aleatorios(pop)
# print(pop)
```

```python
def teste_de_adaptabilidade(indiv):
    #aqui vai a métrica de sobrevivência
    if indiv<8:
        return False #morre
    return True #vive
    
def sobreviventes_apos_selecao(pop):
    individuos_para_morrerem=[]
    for i in range(0,len(pop)):
        indiv=pop[i]
        chance_de_se_safar=random.random()
        if chance_de_se_safar<0.5: # se individuo teve o azar de cair na sorte da seleção natural...
            if (teste_de_adaptabilidade(indiv) == False):#vamos ver ao menos se ele sobrevive ao destino
               individuos_para_morrerem.append(i)
#     print(individuos_para_morrerem)
    sobreviventes = [i for j, i in enumerate(pop) if j not in individuos_para_morrerem]
    return sobreviventes

# sobreviventes=sobreviventes_apos_selecao(pop)
# print(sobreviventes)                    
```

```python
def avancar_uma_geracao(pop):
    pop=cruzar_aleatoriamente(pop)
    pop=sobreviventes_apos_selecao(pop)
    return pop
```

```python
def average(lst): 
    return sum(lst) / len(lst)

def avancar_n_geracoes(pop,n):
    i=0
    while i<n:
        pop=avancar_uma_geracao(pop)
#         print(pop)
        print(average(pop))
        i=i+1
    return pop



# pop_final=avancar_n_geracoes(pop,10)
```
